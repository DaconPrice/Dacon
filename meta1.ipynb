```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 50
```

```python
train_df = pd.read_csv('../data/dacon_veg/train/train.csv')
train_meta_1 = pd.read_csv('../data/dacon_veg/train/meta/TRAIN_산지공판장_2018-2021.csv')
```

## 감자

```python
potato = train_meta_1[(train_meta_1['품목명']=='감자') & (train_meta_1['품종명']== '수미') & (train_meta_1['등급명']== '상')]
potato
```

```python
potato_df = potato.groupby(['시점', '품목명']).mean(numeric_only=True).reset_index()
potato_df
```

## 사과

```python
apple = train_meta_1[(train_meta_1['품목명']=='사과') & (train_meta_1['품종명']== '홍로') & (train_meta_1['등급명']== '상')]
apple_df = apple.groupby(['시점', '품목명']).mean(numeric_only=True).reset_index()
apple_df
```

```python
apple2 = train_meta_1[(train_meta_1['품목명']=='사과') & (train_meta_1['품종명']== '후지') & (train_meta_1['등급명']== '상')]
apple_df2 = apple2.groupby(['시점', '품목명']).mean(numeric_only=True).reset_index()
apple_df2
```

```python
apple_total = pd.concat([apple_df, apple_df2], ignore_index=True)
apple_total
```
